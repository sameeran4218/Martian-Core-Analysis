## Part 1

**Extract amplitude, freqeuncy and phase shift from a given wave form**

```python
import numpy as np
import matplotlib.pyplot as plt
from obspy import read
from scipy.signal import hilbert
from google.colab import drive

file_path='Wave.mseed'  # add a .mseed file here

stream = read(file_path)

# Extract the first trace (assuming single-component seismic data)
trace = stream[0]
sampling_rate = trace.stats.sampling_rate
time_array = np.linspace(0, trace.stats.npts / sampling_rate, trace.stats.npts)

# Compute the amplitude envelope using Hilbert Transform
analytic_signal = hilbert(trace.data)
amplitude_envelope = np.abs(analytic_signal)

# Compute instantaneous phase and frequency
instantaneous_phase = np.unwrap(np.angle(analytic_signal))
instantaneous_frequency = np.gradient(instantaneous_phase) * sampling_rate / (2 * np.pi)

# Compute RMS (Root Mean Square) values
def compute_rms(data):
    return np.sqrt(np.mean(np.square(data)))

rms_amplitude = compute_rms(amplitude_envelope)
rms_frequency = compute_rms(instantaneous_frequency)
rms_phase_shift = compute_rms(instantaneous_phase)

# Print RMS values
print(f"RMS Amplitude: {rms_amplitude:.5f}")
print(f"RMS Frequency: {rms_frequency:.5f} Hz")
print(f"RMS Phase Shift: {rms_phase_shift:.5f} radians")

# Plot the original waveform
plt.figure(figsize=(12, 5))
plt.plot(time_array, trace.data, color="blue", label="Seismic Waveform")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.title("Seismic Waveform")
plt.legend()
plt.show()
```

**Generate a synthetic dataset containing all the features on which the classification of shadow zone depends**

```python
import numpy as np
import pandas as pd
from obspy import UTCDateTime, Stream, Trace

# Re-defining required constants
mars_radius = 3389.5  # Mars radius in km
p_wave_velocity = 6.0  # Approximate velocity in km/s for Mars
s_wave_velocity = 3.5  # Approximate velocity in km/s for Mars

# Adjusting parameters for higher frequency data recording
recording_interval = 15  # Every 15 seconds
total_seconds = 3 * 24 * 3600  # Total duration: 3 days
num_samples = total_seconds // recording_interval  # Number of data points

# Generate timestamps for each recording interval
relative_time = np.arange(0, total_seconds, recording_interval)

# Generate synthetic data for each time step
distance_angle = np.random.uniform(0, 180, num_samples)  # Distance in degrees
instantaneous_amplitude = np.random.uniform(0.1, 10.0, num_samples)  # Arbitrary amplitude scale
instantaneous_frequency = np.random.uniform(0.01, 5.0, num_samples)  # Frequency in Hz
rms_amplitude = np.random.uniform(0.05, 5.0, num_samples)  # RMS amplitude
rms_frequency = np.random.uniform(0.005, 2.5, num_samples)  # RMS frequency
phase_shift = np.random.uniform(0, 360, num_samples)  # Phase shift in degrees
wave_type = np.random.choice(["P", "S"], num_samples)  # P-wave or S-wave
depth_of_source = np.random.uniform(0, 700, num_samples)  # Depth in km

# Compute P-wave and S-wave arrival times
arc_length = (np.pi / 180) * distance_angle * mars_radius
p_wave_arrival_time = relative_time + arc_length / p_wave_velocity
s_wave_arrival_time = relative_time + arc_length / s_wave_velocity

# Generate synthetic P-wave and S-wave velocities
p_wave_velocity_data = np.random.uniform(5.5, 6.5, num_samples)  # P-wave velocity in km/s
s_wave_velocity_data = np.random.uniform(3.0, 4.0, num_samples)  # S-wave velocity in km/s


# Create DataFrame with updated data
df = pd.DataFrame({
    "relative_time": relative_time,
    "distance_angle": distance_angle,
    "instantaneous_amplitude": instantaneous_amplitude,
    "instantaneous_frequency": instantaneous_frequency,
    "rms_amplitude": rms_amplitude,
    "rms_frequency": rms_frequency,
    "phase_shift": phase_shift,
    "wave_type": wave_type,
    "P_wave_arrival_time": p_wave_arrival_time,
    "S_wave_arrival_time": s_wave_arrival_time,
    "depth_of_source": depth_of_source,
    "P_wave_velocity": p_wave_velocity_data,
    "S_wave_velocity":s_wave_velocity_data
})


df.to_csv('synthetic_mars_data.csv',index=False)
df.head()
```

**Label the shadow zones based on using the astronomical aspect of the features**

```python

# Define feature importance weights (assigning weights based on their relevance to shadow zones)
feature_weights = {
    "depth_of_source": 0.3107359667686368,  # Most important
    "distance_angle": 0.2867504460560678,  # Equally important as depth of source
    "rms_amplitude": 0.12729304200973518,  # Less important but contributes
    "P_wave_arrival_time": 0.090618912871697,  # Less important but contributes
    "instantaneous_amplitude": 0,  # Minor influence
    "instantaneous_frequency": 0,  # Minor influence
    "rms_frequency": 0.014571166056313417,  # Minor influence
    "phase_shift": 0,  # Minor influence
    "P_wave_velocity": 0.01208854216460053,  # Minor influence
    "S_wave_velocity": 0   # Minor influence
}

# Normalize weights to sum to 1
total_weight = sum(feature_weights.values())
feature_weights = {k: v / total_weight for k, v in feature_weights.items()}

# Define physics-based thresholds for determining shadow zone
shadow_zone_angle_min = 103  # Degrees
shadow_zone_angle_max = 143  # Degrees
depth_threshold = 500  # Deeper sources more likely in shadow zone
amplitude_threshold = 2.0  # Low amplitude suggests wave absorption in shadow zone
p_wave_arrival_threshold = df["P_wave_arrival_time"].median()  # Use median as a reference

# Compute shadow zone score based on weighted features
shadow_zone_score = (
    feature_weights["depth_of_source"] * (df["depth_of_source"] > depth_threshold).astype(int) +
    feature_weights["distance_angle"] * ((df["distance_angle"] >= shadow_zone_angle_min) & (df["distance_angle"] <= shadow_zone_angle_max)).astype(int) +
    feature_weights["rms_amplitude"] * (df["rms_amplitude"] < amplitude_threshold).astype(int) +
    feature_weights["P_wave_arrival_time"] * (df["P_wave_arrival_time"] > p_wave_arrival_threshold).astype(int) +
    feature_weights["instantaneous_amplitude"] * (df["instantaneous_amplitude"] < amplitude_threshold).astype(int) +
    feature_weights["instantaneous_frequency"] * (df["instantaneous_frequency"] < 1.0).astype(int) +
    feature_weights["rms_frequency"] * (df["rms_frequency"] < 1.0).astype(int) +
    feature_weights["phase_shift"] * (df["phase_shift"] > 180).astype(int) +  # Random assumption: Higher phase shifts correlate with shadow zones
    feature_weights["P_wave_velocity"] * (df["P_wave_velocity"] < 6.0).astype(int) +
    feature_weights["S_wave_velocity"] * (df["S_wave_velocity"] < 3.5).astype(int)
)
df=df.copy()

# Define threshold for classification (e.g., more than 50% weighted score means shadow zone)
df.loc[:, "shadow_zone"] = (shadow_zone_score >= 0.5).astype(int)
df.to_csv('synthetic_labeled_data.csv',index=False)
df.head()
```

```python
import pandas as pd
import numpy as np
from math import radians, degrees, acos, sin, cos

# Function to calculate distance in angle between two coordinates
def calculate_distance_angle(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    return degrees(acos(sin(lat1) * sin(lat2) + cos(lat1) * cos(lat2) * cos(lon1 - lon2)))

# User input
depth_of_source = float(input("Enter depth of source: "))
source_lat = float(input("Enter source latitude: "))
source_lon = float(input("Enter source longitude: "))
rms_amplitude = float(input("Enter RMS amplitude: "))
instantaneous_freq = float(input("Enter instantaneous frequency: "))
P_wave_time = float(input("Enter P-wave arrival time: "))


# Mars station coordinates (InSight SEIS station at Elysium Planitia)
station_lat = 4.502  # Latitude of InSight SEIS
station_lon = 135.623  # Longitude of InSight SEIS

distance_angle=calculate_distance_angle(source_lat, source_lon, station_lat, station_lon)

user_input = np.array([[
    float(distance_angle),
    float(depth_of_source),
    float(rms_amplitude),
    float(instantaneous_freq),
    float(P_wave_time)
]], dtype=np.float64)

# Convert user input into a DataFrame
user_input_df = pd.DataFrame([{
    'depth': float(depth_of_source),
    'distance_angle': float(distance_angle),
    'rms_amplitude': float(rms_amplitude),
    'P_wave_time': float(P_wave_time),
    'inst_freq': float(instantaneous_freq),
}])

user_prediction = rf_random.best_estimator_.predict(user_input_df)
print("Predicted Shadow Zone Status:", "Yes" if user_prediction[0] == 1 else "No")
```

### **mRMR(minimum redundance maximum relevance)**

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_regression
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Features and Target
features = [
    'relative_time', 'distance_angle', 'density','rms_frequency','instantaneous_frequency','P_wave_arrival_time', 'S_wave_arrival_time', 'bulk_modulus', 'shear_modulus',
]
X = df[features]
y = df['core_radius']

# Step 1: Calculate Mutual Information (Relevance)
mi_scores = mutual_info_regression(X, y)

# Step 2: Calculate Redundancy (Correlation Matrix)
correlation_matrix = X.corr().abs()

# mRMR Calculation: Score = Relevance - Redundancy
mrmr_scores = {}
for i, feature in enumerate(features):
    relevance = mi_scores[i]
    redundancy = correlation_matrix[feature].mean()  # Average correlation with other features
    mrmr_score = relevance - redundancy
    mrmr_scores[feature] = mrmr_score

# Selecting Top Features Based on mRMR Scores
selected_features_df = pd.DataFrame(list(mrmr_scores.items()), columns=['Feature', 'mRMR Score'])
selected_features_df = selected_features_df.sort_values(by='mRMR Score', ascending=False)

# Displaying the DataFrame
print(selected_features_df)

# Visualizing the mRMR Scores
plt.figure(figsize=(12, 6))
plt.barh(selected_features_df['Feature'], selected_features_df['mRMR Score'], color='skyblue')
plt.xlabel('mRMR Score')
plt.title('Feature Selection using mRMR')
plt.gca().invert_yaxis()  # To display the highest score on top
plt.show()
```

## **Checking Outliers in the Dataset**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is already loaded
# List of features to check for outliers
features = ['vp', 'vs', 'density', 'core_radius']

# Function to detect outliers using the IQR method
def detect_outliers(df, feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
    return outliers

# Detecting outliers for each feature
outlier_summary = {}
for feature in features:
    outliers = detect_outliers(df, feature)
    outlier_summary[feature] = len(outliers)

# Displaying the number of outliers for each feature
for feature, count in outlier_summary.items():
    print(f"Outliers in {feature}: {count}")

# Visualizing outliers using boxplots
plt.figure(figsize=(15, 10))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=df[feature])
    plt.title(f'Boxplot for {feature}')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

# Define known physical constraints for Mars (values based on literature)
mars_layers = {
    "core": {
        "bulk_modulus": (100, 300),  # GPa
        "shear_modulus": (0, 30),  # GPa (core has low shear modulus)
        "density": (5000, 7000),  # kg/m³
    },
    "mantle": {
        "bulk_modulus": (50, 150),  # GPa
        "shear_modulus": (30, 90),  # GPa
        "density": (3000, 5000),  # kg/m³
    },
    "crust": {
        "bulk_modulus": (10, 70),  # GPa
        "shear_modulus": (10, 50),  # GPa
        "density": (2500, 3500),  # kg/m³
    }
}

# Function to compute P-wave and S-wave velocities
def compute_velocities(bulk_modulus, shear_modulus, density):
    bulk_modulus *= 1e9  # Convert GPa to Pascals
    shear_modulus *= 1e9  # Convert GPa to Pascals
    density = np.maximum(density, 1)  # Avoid division by zero

    vp = np.sqrt((bulk_modulus + 4/3 * shear_modulus) / density)  # P-wave velocity
    vs = np.sqrt(shear_modulus / density)  # S-wave velocity

    vp = np.maximum(vp, 0)  # Ensure non-negative velocities
    vs = np.maximum(vs, 0)

    return vp, vs

# Generate synthetic dataset
num_samples_per_layer = 5000
data = []

for layer, constraints in mars_layers.items():
    for _ in range(num_samples_per_layer):
        bulk_modulus = np.random.uniform(*constraints["bulk_modulus"])
        shear_modulus = np.random.uniform(*constraints["shear_modulus"])
        density = np.random.uniform(*constraints["density"])

        vp, vs = compute_velocities(bulk_modulus, shear_modulus, density)

        data.append([layer, bulk_modulus, shear_modulus, density, vp, vs])

# Create a DataFrame
columns = ["layer", "bulk_modulus", "shear_modulus", "density", "vp", "vs"]
df = pd.DataFrame(data, columns=columns)

df.to_csv('gan_Data.csv', index=False)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import numpy as np
from torch.utils.data import DataLoader, TensorDataset

# Load the dataset
df = pd.read_csv("gan_Data.csv")

# Normalize the data
min_values = df.iloc[:, 1:].min()
max_values = df.iloc[:, 1:].max()
df.iloc[:, 1:] = (df.iloc[:, 1:] - min_values) / (max_values - min_values)

# Convert data to PyTorch tensors
features = torch.tensor(df.iloc[:, 1:].values, dtype=torch.float32)

# Create DataLoader
dataset = TensorDataset(features)
dataloader = DataLoader(dataset, batch_size=64, shuffle=True)

# Function to compute velocities
def compute_velocities(bulk_modulus, shear_modulus, density):
    # Avoid in-place modification by creating copies
    bulk_modulus = bulk_modulus * 1e9
    shear_modulus = shear_modulus * 1e9
    density = torch.clamp(density, min=1) # using torch.clamp to avoid modifying data in place

    vp = torch.sqrt((bulk_modulus + 4/3 * shear_modulus) / density)
    vs = torch.sqrt(shear_modulus / density)
    # using torch.clamp to avoid modifying data in place
    return torch.clamp(vp, min=0), torch.clamp(vs, min=0)

# Function to check layer consistency
def check_layer_consistency(fake_data):
    bulk_modulus, shear_modulus, density = fake_data[:, 0], fake_data[:, 1], fake_data[:, 2]

    # Using .clone() to create copies and avoid modifying in place
    core_condition = (bulk_modulus > 0.4).clone() & (shear_modulus < 0.05).clone() & (density > 0.5).clone()
    mantle_condition = (bulk_modulus > 0.2).clone() & (shear_modulus > 0.05).clone() & (density > 0.3).clone()
    crust_condition = (bulk_modulus < 0.2).clone() & (shear_modulus < 0.05).clone() & (density < 0.3).clone()
    valid_layer = core_condition | mantle_condition | crust_condition
    return torch.mean(valid_layer.float())

# Generator
class Generator(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(Generator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU(),
            nn.Linear(128, output_dim),
            nn.Sigmoid()
        )

    def forward(self, z):
        return self.model(z)

# Discriminator
class Discriminator(nn.Module):
    def __init__(self, input_dim):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU(),
            nn.Linear(128, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.model(x)

# Initialize models
gen = Generator(input_dim=5, output_dim=5)
disc = Discriminator(input_dim=5)

# Optimizers
optimizer_g = optim.Adam(gen.parameters(), lr=0.0002)
optimizer_d = optim.Adam(disc.parameters(), lr=0.0002)

# Loss function
criterion = nn.BCELoss()

# Training the GAN
epochs = 50
# Enable anomaly detection if needed
# torch.autograd.set_detect_anomaly(True)

for epoch in range(epochs):
    for real_data, in dataloader:
        batch_size = real_data.size(0)

        # Train Discriminator
        real_labels = torch.ones(batch_size, 1, requires_grad=False)
        fake_labels = torch.zeros(batch_size, 1, requires_grad=False)

        outputs = disc(real_data)
        d_loss_real = criterion(outputs, real_labels)

        z = torch.randn(batch_size, 5)
        fake_data = gen(z)

        vp, vs = compute_velocities(fake_data[:, 0], fake_data[:, 1], fake_data[:, 2])
        # Avoid in-place operations here
        penalty = torch.mean((vp < 0).float() + (vs < 0).float()).detach()
        layer_consistency = check_layer_consistency(fake_data).detach()
        consistency_penalty = (1 - layer_consistency).detach()

        outputs = disc(fake_data.detach())
        d_loss_fake = criterion(outputs, fake_labels)
        d_loss = d_loss_real + d_loss_fake
        optimizer_d.zero_grad()
        d_loss.backward()
        optimizer_d.step()

        # Train Generator
        outputs = disc(fake_data)
        g_loss = criterion(outputs, real_labels) + penalty + consistency_penalty

        optimizer_g.zero_grad()
        g_loss.backward()
        optimizer_g.step()

    print(f"Epoch [{epoch+1}/{epochs}], D Loss: {d_loss.item():.4f}, G Loss: {g_loss.item():.4f}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Define the neural network class
class PINN(nn.Module):
    def __init__(self):
        super(PINN, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(3, 256),
            nn.LayerNorm(256),
            nn.LeakyReLU(negative_slope=0.01),
            nn.Dropout(0.1),

            nn.Linear(256, 256),
            nn.LayerNorm(256),
            nn.LeakyReLU(negative_slope=0.01),
            nn.Dropout(0.1),

            nn.Linear(256, 256),
            nn.LayerNorm(256),
            nn.LeakyReLU(negative_slope=0.01),
            nn.Dropout(0.1),

            nn.Linear(256, 128),
            nn.LayerNorm(128),
            nn.LeakyReLU(negative_slope=0.01),

            nn.Linear(128, 1)
        )

    def forward(self, rho, K, G):
        inputs = torch.cat((rho, K, G), dim=1)
        return self.net(inputs)


# Physics loss with Normalization
def p_wave_velocity_loss(model, rho, K, G, V_P_actual):
    V_p_pred = model(rho, K, G)
    loss = torch.mean(((V_p_pred - V_P_actual) / V_P_actual.mean()) ** 2)  # Normalized Loss
    return loss

# Supervised loss
def supervised_loss(model, rho, K, G, V_P_True):
    V_p_pred = model(rho, K, G)
    return torch.mean((V_p_pred - V_P_True) ** 2)  # MSE loss

# Training setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
pinn = PINN().to(device)
optimizer = optim.AdamW(pinn.parameters(), lr=0.001, weight_decay=1e-6)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=500, verbose=True)

# Move and standardize data
K_train = torch.tensor(X_train["bulk_modulus"].values * (10**9), dtype=torch.float32).view(-1, 1).to(device)
G_train = torch.tensor(X_train["shear_modulus"].values * (10**9), dtype=torch.float32).view(-1, 1).to(device)
rho_train = torch.tensor(X_train["density"].values, dtype=torch.float32).view(-1, 1).to(device)
V_P_True = torch.tensor(y_train["vp"].values, dtype=torch.float32).view(-1, 1).to(device)

V_P_actual = torch.sqrt((K_train + (4/3) * G_train) / rho_train).to(device)

# Feature Standardization
K_mean, K_std = K_train.mean(), K_train.std()
G_mean, G_std = G_train.mean(), G_train.std()
rho_mean, rho_std = rho_train.mean(), rho_train.std()

K_train = (K_train - K_mean) / K_std
G_train = (G_train - G_mean) / G_std
rho_train = (rho_train - rho_mean) / rho_std

# Training loop
num_epochs = 10000
lambda_supervised = 0.1
best_loss = float("inf")

for epoch in range(num_epochs):
    optimizer.zero_grad()

    physics_loss = p_wave_velocity_loss(pinn, rho_train, K_train, G_train, V_P_actual)
    data_loss = supervised_loss(pinn, rho_train, K_train, G_train, V_P_True)

    loss = physics_loss + lambda_supervised * data_loss

    loss.backward()
    optimizer.step()
    scheduler.step(loss)

    if epoch % 500 == 0:
        print(f"Epoch {epoch}, Total Loss: {loss.item()}, Physics Loss: {physics_loss.item()}, Supervised Loss: {data_loss.item()}")

# Predicting Vp
def predict_Vp(rho, K, G):
    rho_tensor = torch.tensor([[rho]], dtype=torch.float32).to(device)
    K_tensor = torch.tensor([[K * 10**9]], dtype=torch.float32).to(device)
    G_tensor = torch.tensor([[G * 10**9]], dtype=torch.float32).to(device)

    # Standardize Inputs
    K_tensor = (K_tensor - K_mean) / K_std
    G_tensor = (G_tensor - G_mean) / G_std
    rho_tensor = (rho_tensor - rho_mean) / rho_std

    with torch.no_grad():
        V_p_pred = pinn(rho_tensor, K_tensor, G_tensor)

    return V_p_pred.item()
```

## Predicting S wave velocity

```python
import pandas as pd
df = pd.read_csv('synthetic_data.csv')
from sklearn.model_selection import train_test_split
X=df.drop(['vp','vs'],axis=1)
y=df[['vs']]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Define the neural network class
class PINN(nn.Module):
    def __init__(self):
        super(PINN, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(3, 256),
            nn.LeakyReLU(negative_slope=0.01),
            nn.Dropout(0.1),

            nn.Linear(256, 256),
            nn.LeakyReLU(negative_slope=0.01),
            nn.Dropout(0.1),

            nn.Linear(256, 256),
            nn.LeakyReLU(negative_slope=0.01),
            nn.Dropout(0.1),

            nn.Linear(256, 128),
            nn.LeakyReLU(negative_slope=0.01),

            nn.Linear(128, 1)  # Output: V_s (S-wave velocity)
        )

    def forward(self, rho, K, G):
        inputs = torch.cat((rho, K, G), dim=1)
        return self.net(inputs)

# Physics loss enforcing the S-wave velocity equation
def s_wave_velocity_loss(model, rho, G, V_S_actual):
    V_s_pred = model(rho, torch.zeros_like(G), G)  # Bulk modulus is irrelevant for Vs
    return torch.mean((V_s_pred - V_S_actual) ** 2)  # MSE loss

# Supervised loss using real data
def supervised_loss(model, rho, K, G, V_S_True):
    V_s_pred = model(rho, K, G)
    return torch.mean((V_s_pred - V_S_True) ** 2)  # MSE loss

# Training setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
pinn = PINN().to(device)
optimizer = optim.AdamW(pinn.parameters(), lr=0.001, weight_decay=1e-6)
scheduler = torch.optim.lr_scheduler.CyclicLR(optimizer, base_lr=1e-4, max_lr=1e-3, step_size_up=2000, mode="triangular")

# Move data to device
G_train = torch.tensor(X_train["shear_modulus"].values * (10**9), dtype=torch.float32).view(-1, 1).to(device)
rho_train = torch.tensor(X_train["density"].values, dtype=torch.float32).view(-1, 1).to(device)
V_S_True = torch.tensor(y_train["vs"].values, dtype=torch.float32).view(-1, 1).to(device)
V_S_actual = torch.sqrt(G_train / rho_train).to(device)

# Standardize Inputs
G_mean, G_std = G_train.mean(), G_train.std()
rho_mean, rho_std = rho_train.mean(), rho_train.std()

G_train = (G_train - G_mean) / G_std
rho_train = (rho_train - rho_mean) / rho_std

# Training loop
num_epochs = 10000
lambda_supervised = 0.1
for epoch in range(num_epochs):
    optimizer.zero_grad()
    physics_loss = s_wave_velocity_loss(pinn, rho_train, G_train, V_S_actual)
    data_loss = supervised_loss(pinn, rho_train, torch.zeros_like(G_train), G_train, V_S_True)
    loss = physics_loss + lambda_supervised * data_loss  # Total loss
    loss.backward()
    optimizer.step()
    scheduler.step()

    if epoch % 500 == 0:
        print(f"Epoch {epoch}, Total Loss: {loss.item()}, Physics Loss: {physics_loss.item()}, Supervised Loss: {data_loss.item()}")

# Predicting V_s for new inputs
def predict_Vs(rho, G):
    rho_tensor = torch.tensor([[rho]], dtype=torch.float32, requires_grad=True).to(device)
    G_tensor = torch.tensor([[G * 10**9]], dtype=torch.float32, requires_grad=True).to(device)

    # Standardize Inputs
    G_tensor = (G_tensor - G_mean) / G_std
    rho_tensor = (rho_tensor - rho_mean) / rho_std

    V_s_pred = pinn(rho_tensor, torch.zeros_like(G_tensor), G_tensor)
    return V_s_pred.item()
```

```python
# Example prediction
rho_test, G_test = 6742.56, 4.224  # Keep K and G in GPa
Vs_prediction = predict_Vs(rho_test, G_test)
print(f"Predicted V_p: {Vs_prediction}")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import torch

# Move test data to device
G_test = torch.tensor(X_test["shear_modulus"].values * (10**9), dtype=torch.float32).view(-1, 1).to(device)
rho_test = torch.tensor(X_test["density"].values, dtype=torch.float32).view(-1, 1).to(device)

# Standardize test data using training statistics
G_test = (G_test - G_mean) / G_std
rho_test = (rho_test - rho_mean) / rho_std

# Predict V_s
with torch.no_grad():
    V_s_pred = pinn(rho_test, torch.zeros_like(G_test), G_test).cpu().numpy()

# Get true V_s values
V_s_true = y_test["vs"].values  # Assuming y_test contains the true S-wave velocities

# Compute evaluation metrics
mse = mean_squared_error(V_s_true, V_s_pred)
mae = mean_absolute_error(V_s_true, V_s_pred)
r2 = r2_score(V_s_true, V_s_pred)

# Print results
print(f"Evaluation Metrics on X_test:")
print(f"Mean Squared Error (MSE): {mse:.6f}")
print(f"Mean Absolute Error (MAE): {mae:.6f}")
print(f"R² Score: {r2:.6f}")
```
